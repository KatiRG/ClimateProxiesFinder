```python
#Script to optimize climate proxy dataset
#Converts strings in Archive, Material, DOI and Reference columns to ints using dicts

#Created: 24.02.2016
#Last modified: 
```

```python
import numpy as np
import pandas as pd
import collections
import os
```

```python
# Create dataframe from analogues file

df = pd.DataFrame()
df = pd.read_csv('proxies.tsv', delimiter='\t', index_col=0)
#df.set_index('Date', inplace=True)

df.head(5)

#dfS.index.name
```

```python
df['DOI'].unique().shape
#df['DOI'].shape
```

```python
#MAKE DICTIONARIES

#DOI dict

DOIarr = df['DOI'].unique()
dfDOI = pd.DataFrame()
dfDOI['DOI'] = DOIarr   #.toarray().tolist()
dfDOI.index.name = 'id'
dfDOI

#make a dict from DOI df
#DOI_dict = dfDOI.to_dict()
#DOI_dict

#s = dfDOI.apply((lambda x: x.to_dict()), axis=1)
#s

DOI_dict = dict(zip(dfDOI.DOI, dfDOI.index))
DOI_dict
```

```python
#Reference dict
REFarr = df['Reference'].unique()
dfREF = pd.DataFrame()
dfREF['REF'] = REFarr
dfREF.index.name = 'id'

REF_dict = dict(zip(dfREF.REF, dfREF.index))
#REF_dict
```

```python
#Archive and Material dicts

archive_dict={
        u"Ice": 1,
        u"Lake": 2,
        u"Ocean": 3,
        u"Speleothem": 4,
        u"Tree": 5
}

material_dict={
        u"Benthic foraminifera": 1,
        u"Carbonate": 2,
        u"Cellulose": 3,
        u"Coral": 4,
        u"Ice": 5,
        u"Non-Carbonate": 6,
        u"Others": 7,
        u"Planktonic foraminifera": 8,
        u"Speleothem": 9,
        u"Unknown": 10
    
}
```

```python
#Replace Archive, Material, DOI and Reference columns by their dictionary keys

df_opt_all = df.replace({'Archive': archive_dict, 'Material': material_dict, 'DOI': DOI_dict, 'Reference': REF_dict})
df_opt_all.head(10)
```
