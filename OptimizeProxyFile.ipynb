```python
import numpy as np
import pandas as pd
import collections
import os
```

```python
# Create dataframe from analogues file

df = pd.DataFrame()
df = pd.read_csv('proxies.tsv', delimiter='\t', index_col=0)
#df.set_index('Date', inplace=True)

df.head(5)

#dfS.index.name
```

```python
#drop DOI and Reference columns for now
#DF.drop('column_name', axis=1, inplace=True)
#DF.drop(DF.columns[[0, 1, 3]], axis=1) # Note: zero indexed

df.drop('DOI', axis=1, inplace=True)
df.drop('Reference', axis=1, inplace=True)

df.head(10)
```

```python
#Replace Archive and Material strings by number identifiers

archive_dict={
        u"Ice": 1,
        u"Lake": 2,
        u"Ocean": 3,
        u"Speleothem": 4,
        u"Tree": 5
}

material_dict={
        u"Benthic foraminifera": 1,
        u"Carbonate": 2,
        u"Cellulose": 3,
        u"Coral": 4,
        u"Ice": 5,
        u"Non-Carbonate": 6,
        u"Others": 7,
        u"Planktonic foraminifera": 8,
        u"Speleothem": 9,
        u"Unknown": 10
    
}

df_opt = df.replace({'Archive': archive_dict, 'Material': material_dict})
```
